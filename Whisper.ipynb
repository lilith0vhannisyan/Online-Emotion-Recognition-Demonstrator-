```python
import os

# Use a relative path from the notebook's location
folder_path = '../IS2025_Podcast_Challenge/Audios'

# Check if the folder exists
if os.path.exists(folder_path):
    # List all files in the folder
    files = os.listdir(folder_path)

    # Filter and print the first 10 file names
    for file in files[:10]:
        print(file)
else:
    print(f"Folder does not exist: {folder_path}")
```

```python
import os
import whisper

# Load the Whisper model
model = whisper.load_model("base")  # Change model size if needed: "tiny", "small", "medium", "large", "large-v2"

def transcribe_audio_files(audio_directory, output_directory):
    # Creating output directory if it doesn't exist
    os.makedirs(output_directory, exist_ok=True)

    # Iterate over all audio files in the directory
    for filename in os.listdir(audio_directory):
        if filename.endswith(".wav") or filename.endswith(".mp3") or filename.endswith(".m4a"):
            file_path = os.path.join(audio_directory, filename)
            print(f"Processing: {file_path}")

            try:
                # Perform the transcription
                result = model.transcribe(file_path)
                transcription = result["text"]

                # Save the transcription to a file
                output_filename = os.path.splitext(filename)[0] + ".txt"
                output_path = os.path.join(output_directory, output_filename)
                with open(output_path, "w") as output_file:
                    output_file.write(transcription)
                print(f"Saved transcription to: {output_path}")
            except Exception as e:
                print(f"Error transcribing {file_path}: {e}")

# Define the paths 
audio_directory = "../IS2025_Podcast_Challenge/Audios"  
output_directory = "MyTranscriptions"  # Path to save transcriptions

# Transcribe the audio files
transcribe_audio_files(audio_directory, output_directory)
```

```python
import os
import whisper
from IPython.display import Audio

# Load the Whisper model
model = whisper.load_model("base")  # Change model size if needed: "tiny", "small", "medium", "large", "large-v2"

def transcribe_and_listen(audio_directory):
    # Iterate over all audio files in the directory
    for filename in os.listdir(audio_directory):
        if filename.endswith(".wav") or filename.endswith(".mp3") or filename.endswith(".m4a"):
            file_path = os.path.join(audio_directory, filename)
            print(f"Processing: {file_path}")

            try:
                # Perform the transcription
                result = model.transcribe(file_path)
                transcription = result["text"]

                # Print the transcription
                print(f"Transcription of {filename}:")
                print(transcription)

                # Play the audio file
                print(f"Playing the audio of {filename}:")
                display(Audio(file_path))  # This will play the audio directly in the notebook
                
                break  # Only process the first file and then stop
            except Exception as e:
                print(f"Error transcribing {file_path}: {e}")

# Define the path to the audio directory
audio_directory = "../IS2025_Podcast_Challenge/Audios"  # Replace with your path

# Transcribe the first audio file and print the transcription, then play it
transcribe_and_listen(audio_directory)
```
